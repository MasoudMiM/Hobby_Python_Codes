```python
%pylab inline
import math
import numpy as np


## Brownian Time
eta=40e-3 # Viscosity [Pa.s] for EMG 304
k=1.3806488e-23 #Boltzman's constant [m^2 kg s^(-2) K^(-1)]
T=23+274.15 # Temperature [Kelvin]
r=np.linspace(0,20,200,endpoint=False)*1e-9
To_B=[]


for i in range(len(r)):
    To_B.append(np.divide((4*np.pi*eta*r[i]**3),(k*T)))

thefont = {'fontname':'Helvetica'}

pylab.plot(r*1e9,To_B,lw=3,label=r'$\tau_B$')
pylab.ylabel(r'$\tau\ (s)$',fontsize=20)
pylab.xlabel(r'$Radius\ (nm)$',fontsize=20)
gf = pylab.gcf()
ax = gf.add_subplot(111)
ax.set_yscale("log")

## Neel Time
Tau_0=1e-9
K=1.1e4 # anisotropy constant [kJ/m^3] for magnetite
V=[]
To_N=[]

for i in range(len(r)):
    V.append(4./3*np.pi*r[i]**3)
    To_N.append(Tau_0*np.exp(K*V[i]/(k*T)))

#pylab.figure()
pylab.plot(r*1e9,To_N,lw=3,label=r'$\tau_N$')
pylab.ylabel(r'$\tau\ (s)$',fontsize=20)
pylab.xlabel(r'$Radius\ (nm)$',fontsize=20)
gf = pylab.gcf()
ax = gf.add_subplot(111)
ax.set_yscale("log")


## Effetive relaxation time
To_eff=np.divide((np.multiply(To_B,To_N)),np.add(To_B,To_N))

#pylab.figure()
pylab.plot(r*1e9,To_eff,lw=3,label=r'$\tau_{eff}$')
pylab.ylabel(r'$\tau\ (s)$',fontsize=20)
pylab.xlabel('Radius (nm)',fontsize=20)
gf = pylab.gcf()
ax = gf.add_subplot(111)
ax.set_yscale("log")
gf.set_size_inches(10,8)
pylab.grid(color='black', alpha=0.6, linestyle='dashed', linewidth=0.5)
pylab.grid(color='black', alpha=0.6, linestyle='dashed', linewidth=0.5, which='minor')
pylab.legend(loc='best')
```

```python
## Zoomed figure

pylab.figure()
fig_zoom=pylab.gcf()
ax = fig_zoom.add_subplot(111)
pylab.plot(r[75:130]*1e9,To_B[75:130],lw=5,label=r'$\tau_B$')
pylab.plot(r[75:130]*1e9,To_N[75:130],lw=5,label=r'$\tau_N$')
pylab.plot(r[75:130]*1e9,To_eff[75:130],lw=5,label=r'$\tau_{eff}$')
ax.set_yscale("log")
pylab.ylabel(r'$\tau\ (s)$',fontsize=40)
pylab.xlabel('Radius (nm)',fontsize=40)
fig_zoom.set_size_inches(10,8)
pylab.xlim([7.5, 12.9])
pylab.grid(color='black', alpha=0.6, linestyle='dashed', linewidth=0.5)
pylab.grid(color='black', alpha=0.6, linestyle='dashed', linewidth=0.5, which='minor')
pylab.legend(loc='best')
```

```python
## Imaginary part of the magnetic susceptibility
X0=5.03 # Magnetic susceptibility for EMG 304 (SI)
rad=np.multiply([5,7,9,11,13,15],1e-9)
To_BB=[]
Vol=[]
To_NN=[]
To_total=[]

for i in range(len(rad)):
    To_BB.append(np.divide((4*np.pi*eta*rad[i]**3),(k*T)))
    Vol.append(4./3*np.pi*rad[i]**3)
    To_NN.append(Tau_0*np.exp(K*Vol[i]/(k*T)))
    To_total.append(np.add(To_BB,To_NN))
    
frq=np.linspace(10,10e4,2000) # Frequency range
w=np.multiply(frq,np.pi)

pylab.figure()

for j in range(len(To_total)):
    X=[]
    for i in range(len(w)):
        # Only Brownian relaxation time is considered
        X.append(np.divide((np.multiply(np.multiply(w[i],To_BB[j]),X0)),\
                           (np.add(1,np.multiply(np.multiply(w[i],To_BB[j]),np.multiply(w[i],To_BB[j]))))))
    
    pylab.plot(frq,X,lw=3)
    
pylab.ylabel("Imaginary component of susceptibility",fontsize=20)
pylab.xlabel("Frequency (Hz)",fontsize=20)
gf = pylab.gcf()
ax = gf.add_subplot(111)
ax.set_xscale("log")
gf.set_size_inches(10,8)
pylab.grid(color='b', alpha=0.6, linestyle='dashed', linewidth=0.5)
pylab.grid(color='b', alpha=0.6, linestyle='dashed', linewidth=0.5, which='minor')
```
