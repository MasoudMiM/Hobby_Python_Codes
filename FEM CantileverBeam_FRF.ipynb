```python
%pylab inline
import math
import numpy as np
import scipy
from scipy.fftpack import fft

import warnings
warnings.simplefilter('ignore',DeprecationWarning);

## ---------- Given values
#      |
#      |= = = = = = ... ====|    <= Beam structure  
#      |1,2,3,4,5,6,...,..,n     <= # of elements

n=7;  # Number of elements   
l=0.7; # Length of the beam
ro=7850; # Density
A=100e-6; # Cross-section area
E=200e9;
I=833.3333333333e-12; # Moment of area (I)
le = np.zeros(n)
le[:] = np.divide(l,n);

## ---------- Mass and stiffness matrices for elements
me = np.zeros([n,4,4]);
ke = np.zeros([n,4,4]);

for i in range(n):
    
    me[i,::]=(ro*A*le[i]/420.)*np.array([[156,22*le[i],54,-13*le[i]],[22*le[i],4*le[i]**2,13*le[i],-3*le[i]**2],
                     [54,13*le[i],156,-22*le[i]],[-13*le[i],-3*le[i]**2,-22*le[i],4*le[i]**2]]);
    
    ke[i,::]=(E*I/le[i]**3)*np.array([[12,6*le[i],-12,6*le[i]],[6*le[i],4*le[i]**2,-6*le[i],2*le[i]**2],
                                     [-12,-6*le[i],12,-6*le[i]],[6*le[i],2*le[i]**2,-6*le[i],4*le[i]**2]]);
    

## ---------- Assemblying procedure
nodes = n+1;
dof = 2*nodes;
K = np.vstack([np.hstack([ke[0,::],np.zeros([4,dof-4])]),np.zeros([dof-4,dof])]);
M = np.vstack([np.hstack([me[0,::],np.zeros([4,dof-4])]),np.zeros([dof-4,dof])]);

for i in range(n-1):
    K = K + np.vstack([np.zeros([2*(i+1),dof]),np.hstack([np.zeros([4,2*(i+1)]), ke[i+1,::], np.zeros([4,dof-4-(2*(i+1))])]),np.zeros([dof-4-(2*(i+1)),dof])]);
    M = M + np.vstack([np.zeros([2*(i+1),dof]),np.hstack([np.zeros([4,2*(i+1)]), me[i+1,::], np.zeros([4,dof-4-(2*(i+1))])]),np.zeros([dof-4-(2*(i+1)),dof])]);


# ================================ B.C. ===========================
# ---------- Boundary Conditions (Both side springs)
#k_b = 1000; # Stiffness of spring for mounting

#K[0,0]=K[0,0]+k_b;
#K[dof-2,dof-2]=K[dof-1,dof-1]+k_b;

# ---------- Boundary Conditions (Cantilever)
K =np.delete(K, (0,1), axis=0);
K =np.delete(K, (0,1), axis=1);
M =np.delete(M, (0,1), axis=0);
M =np.delete(M, (0,1), axis=1);
dof = dof-2;

# ================= Plotting Frequency Response Functions ==================
def PlottingFRF(omg,FPoint,RPoint):
    
    Freq = np.divide(omg,2*np.pi);
    Alpha = np.zeros([len(omg),dof,dof]);
    
    for i in range(len(omg)):
        Alpha[i,::] = np.linalg.inv(np.array([K - omg[i]**2*M]));
    
    Respp = np.absolute(Alpha[:,ForcePoint,ResponsePoint]);
    
    return np.array(Respp), np.array(Freq)


omg = linspace(0,200,300); # Frequency range for FRF (rad/s)

## ------------- force applied on node 2, response measured on node 2 (point FRF)
ForcePoint = 2; # Point for applied force
ResponsePoint = 2; # Point for measuring reponse

Resp1 ,Freq1 = PlottingFRF(omg,ForcePoint,ResponsePoint);

pylab.rcParams.update({'font.size': 20, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
pylab.figure(figsize=(12, 6))

pylab.plot(Freq1, 20*np.log10(Resp1),label=r"$FRF_{2,2}$",linewidth=2);
pylab.ylabel(r"$Receptance\ (dB\ re\ 1 m/N)$")
pylab.xlabel(r"$Frequency\ (Hz)$")

## ------------- force applied on node 2, response measured on node 3
ForcePoint = 2; # Point for applied force
ResponsePoint = 3; # Point for measuring reponse

Resp2 ,Freq2 = PlottingFRF(omg,ForcePoint,ResponsePoint);

pylab.plot(Freq2, 20*np.log10(Resp2),label=r"$FRF_{2,3}$",linewidth=2),pylab.grid(),pylab.xlim([0,Freq2[-1]]);
pylab.legend(loc="best", fontsize=20);
```
