```python
## Importing essential moduli
%pylab inline
import io
import os
from PIL import Image
import numpy as np
from scipy import fftpack
import urllib2
import IPython

import warnings
warnings.simplefilter('ignore',DeprecationWarning);

## Image url
image_url='http://i.imgur.com/8vuLtqi.png'

## Defining essnetial functions
def get_image_from_url(image_url='http://i.imgur.com/8vuLtqi.png', size=(128, 128)):
    file_descriptor = urllib2.urlopen(image_url)
    image_file = io.BytesIO(file_descriptor.read())
    image = Image.open(image_file)
    img_color = image.resize(size, 1)
    img_grey = img_color.convert('L')
    img = np.array(img_grey, dtype=np.float)
    return img
 
def get_2D_dct(img):
    #Get 2D Cosine Transform of Image
    return fftpack.dct(fftpack.dct(img.T, norm='ortho').T, norm='ortho')

def get_2d_idct(coefficients):
    #Get 2D Inverse Cosine Transform of Image
    return fftpack.idct(fftpack.idct(coefficients.T, norm='ortho').T, norm='ortho')

def get_reconstructed_image(raw):
    # Given an interval, values outside the interval are clipped to
    # the interval edges. For example, if an interval of [0, 1] is specified, values
    # smaller than 0 become 0, and values larger than 1 become 1.    
    img = raw.clip(0, 255) 
    img = img.astype('uint8')
    img = Image.fromarray(img)
    return img

## Importing the image and doing DCT
pixels = get_image_from_url(image_url=image_url, size=(256, 256))
dct_size = pixels.shape[0]
dct = get_2D_dct(pixels)
reconstructed_images = []

for ii in range(dct_size):
    dct_copy = dct.copy()
    dct_copy[ii:,:] = 0
    dct_copy[:,ii:] = 0
    
    # Reconstructed image
    r_img = get_2d_idct(dct_copy);
    reconstructed_image = get_reconstructed_image(r_img);

    # Create a list of images
    reconstructed_images.append(reconstructed_image);
    
# Plotting
pylab.figure(figsize=(16, 12));
pylab.grid(b=True, which='minor', linestyle='--')
pylab.grid(b=True, which='major', linestyle='--')
pylab.scatter(range(dct.ravel().size), np.log10(np.abs(dct.ravel())), c='#348ABD', alpha=.3);
pylab.title(r'$DCT\ Coefficient\ Amplitude\ vs.\ Order\ of\ Coefficient$', fontsize=20);
pylab.xlabel(r'$Order\ of\ DCT\ Coefficients$', fontsize=15);
pylab.ylabel(r'$DCT\ Coefficients\ Amplitude\ in\ log\ scale$', fontsize=15);
```

```python
pylab.figure(figsize=(16, 12))
pylab.hist(np.log10(np.abs(dct.ravel())), bins=100, color='#348ABD', alpha=.5, histtype='stepfilled');
pylab.grid(b=True, which='minor', linestyle='--')
pylab.grid(b=True, which='major', linestyle='--')
pylab.xlabel(r'$Amplitude\ of\ DCT\ Coefficients\ (log-scaled)$', fontsize=20);
pylab.ylabel(r'$Number\ of\ DCT\ Coefficients$', fontsize=20);
pylab.title(r'$Amplitude\ distribution\ of\ DCT\ Coefficients$', fontsize=20);
```
