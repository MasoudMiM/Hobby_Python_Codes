```python
## Importing essential moduli
%pylab inline
import io
import os
from PIL import Image
import numpy as np
from scipy import fftpack
import urllib2
import IPython

def get_image_from_url(image_url, size=(128, 128)):
    file_descriptor = urllib2.urlopen(image_url)
    image_file = io.BytesIO(file_descriptor.read())
    image = Image.open(image_file)
    img_color = image.resize(size, 1)
    img_grey = img_color.convert('L')
    img = np.array(img_grey, dtype=np.float)
    return img

## Importing the image and doing DCT
image_url='http://i60.tinypic.com/icnwpy.jpg'
img = get_image_from_url(image_url, size=(256, 256))
pylab.imshow(img, cmap=pylab.cm.gray)

## Discrete Cousine Transform
dct_size = img.shape[0]
# Get 2D Cosine Transform of Image
dct_img = fftpack.dct(fftpack.dct(img.T, norm='ortho').T, norm='ortho')
# Get 2D Inverse Cosine Transform of Image
idct_img = fftpack.idct(fftpack.idct(dct_img.T, norm='ortho').T, norm='ortho')
```

```python
## Plotting Histograms

pylab.figure(figsize=(15, 6))
# Histrogram of the image
subplot(121)
pylab.hist(img.ravel(), bins=254, color='#348ABD', alpha=.5, histtype='stepfilled');
pylab.xlim([0, np.max(np.max(img))])
pylab.xlabel(r'$Amplitude $', fontsize=12);
pylab.ylabel(r'$Number\ of\ occurrence$', fontsize=12);
pylab.title(r'$Amplitude\ distribution$', fontsize=12);

# Histrogram of DCT coefficients
subplot(122)
pylab.hist(np.log10(np.abs(dct_img.ravel())), bins=100, color='#348ABD', alpha=.5, histtype='stepfilled');
pylab.xlabel(r'$Amplitude\ of\ DCT\ Coefficients\ (log-scaled)$', fontsize=12);
pylab.ylabel(r'$Number\ of\ DCT\ Coefficients$', fontsize=12);
pylab.title(r'$Amplitude\ distribution\ of\ DCT\ Coefficients$', fontsize=12);
```

```python
## DCT Coefficients in grid shape
pylab.matshow(np.abs(dct_img), cmap=pylab.cm.Paired);
pylab.grid()
pylab.title(r'$Coefficients\ in\ grid$', fontsize=15);

pylab.matshow(np.abs(dct_img[:50, :50]), cmap=pylab.cm.Paired);
pylab.grid()
pylab.title(r'$First\ 2500\ coefficients\ in\ grid$', fontsize=15);
```

```python
## Applying some functions

# Inverting image
Mx = np.max(np.max(img))
img_1 = np.subtract(Mx,img);
pylab.figure(figsize=(10,6))
subplot(241)
pylab.imshow(img_1, cmap=pylab.cm.gray)
pylab.xticks([]);
pylab.yticks([]);
# Histogram
subplot(245)
pylab.hist(img_1.flatten(),bins=254, color=None, histtype='stepfilled')
xlim([0, 254])

# Converting to some interval [L1,...,L2+L1]
L1 = 50
L2 = 100
img_2 = np.multiply(np.divide(L2,Mx),img)+L1;
subplot(242)
pylab.imshow(img_2, cmap=pylab.cm.gray)
pylab.xticks([]);
pylab.yticks([]);
# Histogram
subplot(246)
pylab.hist(img_2.flatten(),bins=254, color=None, histtype='stepfilled')
xlim([0, 254])

# Squaring the values
img_3 = np.multiply(Mx, np.power(np.divide(img,Mx),2));
subplot(243)
pylab.imshow(img_3, cmap=pylab.cm.gray)
pylab.xticks([]);
pylab.yticks([]);
# Histogram
subplot(247)
pylab.hist(img_3.flatten(),bins=254, color=None, histtype='stepfilled')
xlim([0,254]), ylim([0, 2000])

# Gamma correction
C = 0.9
Gam = 0.3
img_4 = C*np.power(img,Gam)
subplot(244)
pylab.imshow(img_4, cmap=pylab.cm.gray)
pylab.xticks([]);
pylab.yticks([]);
subplot(248)
pylab.hist(img_4, bins=254, color=None,  histtype='stepfilled');
xlim([0, np.max(np.max(img_4))])
```
