```python
%pylab inline
import math
import numpy as np
import sympy
from scipy import integrate
from scipy.fftpack import fft

# Given values
alpha = 7.92e6
m = 1.54
k = 6450.8
c = 0.19
F0 = 3*9.81
f = 10 # frequency [Hz]
w = 2*np.pi*f # frequency [rad/s]


# Defining the right hand side set of eqautions
def rhs(t, X):
    """ RHS of the system """
    x1 = X[0]
    x2 = X[1]
    
    dx1dt = x2
    dx2dt = F0/m*np.cos(w*t)-c/m*x2-k/m*x1-alpha*x1**3
    
    return np.array([dx1dt, dx2dt])

def ODE45Integrate(X0, dt, tmax):
    # Integration!
    
    r = integrate.ode(rhs).set_integrator("dopri5", atol=1.e-10, rtol=1.e-10, nsteps = 200)
    t = 0.0
    r.set_initial_value(X0, t)

    tout = [t]
    x1out = [X0[0]]
    x2out = [X0[1]]
    
    while r.successful() and r.t < tmax:
        r.integrate(r.t + dt)      
               
        tout.append(r.t)
        x1out.append(r.y[0])
        x2out.append(r.y[1])

    return np.array(tout), np.array(x1out), np.array(x2out)
    
X0 = np.array([0.0, 0.0])
tmax = 5.
dt = 4.e-4

t, x1, x2 = ODE45Integrate(X0, dt, tmax)

pylab.rcParams.update({'font.size': 20, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
pylab.figure(figsize=(14, 10))
pylab.subplot(211)
pylab.plot(t, x1, linewidth=2),pylab.xlim([0,5]),pylab.grid()
pylab.ylabel(r"$Displacement\ (m)$")
subplot(212)
pylab.plot(t, x2, linewidth=2),pylab.xlim([0,5]),pylab.grid()
pylab.ylabel(r"$Velocity\ (m/s)$")
pylab.xlabel(r"$Time (s)$")
```

```python
## Transferring to the Fourier domain

npts = len(t)

# Forward transform: f(x) -> F(k)                                                                       
fk = numpy.fft.rfft(x1)
                                    
norm = 2.2/npts

fk = fk*norm
                                          
k = numpy.fft.rfftfreq(npts)
                                                           
kfreq = k*npts/(max(t) + t[1])

# Inverse transform: F(k) -> f(x) -- without the normalization 
fkinv = numpy.fft.irfft(fk/norm)

pylab.rcParams.update({'font.size': 20, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
pylab.figure(figsize=(14,10))
pylab.subplot(211)
pylab.plot(kfreq, numpy.abs(fk), linewidth=2)
pylab.xlabel(r"$Freq\ (Hz)$", fontsize=20), pylab.xlim([0, 20])
pylab.ylabel(r"|X(s)|")
pylab.grid()
pylab.subplot(212)
pylab.plot(t, x1, linewidth=2, ls='o', marker='*', markersize=10, color="red", label="Function"),pylab.xlim([0,1]),pylab.grid()
pylab.plot(t, fkinv.real, label="Inverted Function", linewidth=2, color="green"), pylab.xlim([0, 1])
pylab.xlabel(r"$Time\ (s)$")
pylab.ylabel(r"$x\ (t)$")
pylab.legend(loc="best", fontsize=15)
```
